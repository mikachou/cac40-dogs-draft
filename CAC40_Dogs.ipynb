# functions

```python
def get_stock_price_end(id, year):
    prices = yf.download(id, start="{}-12-24".format(year), end="{}-12-31".format(year), progress=False)
    return prices.iloc[-1]['Close'] if len(prices) > 0 else 0
```

```python
def get_dividends_history(id, year):
    ticker = yf.Ticker(id)
    ticker.history(start="{}-01-01".format(year), end="{}-12-31".format(year))
    return ticker.dividends
```

```python
def get_dividends(id, year):
    return get_dividends_history(id, year).sum()
```

```python
# get tickers from cac40 2022 stocks, which where already presents in 2021
tickers_str = 'ai.pa air.pa alo.pa mt.pa cs.pa bnp.pa en.pa cap.pa ca.pa '\
            + 'aca.pa bn.pa dsy.pa engi.pa el.pa rms.pa ker.pa or.pa lr.pa '\
            + 'mc.pa ml.pa ora.pa ri.pa pub.pa rno.pa saf.pa sgo.pa san.pa '\
            + 'su.pa gle.pa stmpa.pa tep.pa ho.pa tte.pa urw.pa vie.pa dg.pa '\
            + 'viv.pa wln.pa'
# how many stocks?
print(len(tickers_str.split()))
```

```python
def get_actions(id, year):
    ticker = yf.Ticker(id)
    ticker.history(start="{}-01-01".format(year), end="{}-12-31".format(year))
    return ticker.actions
```

```python
keys = tickers_str.upper().split()
keys
```

```python
for key in keys:
    print("key:", key)
    print(get_actions(key, 2021))
    print('-------------------------------------------------------------------')
```

```python
def get_dividends_ss(id, year):
    actions = get_actions(id, year)
    total_dividends = 0
    for idx, action in actions.iterrows():
        if action['Dividends'] > 0:
            total_dividends += action['Dividends']
        elif action['Stock Splits'] > 0:
            total_dividends /= action['Stock Splits']

    return total_dividends
```

```python
for key in keys:
    print("key:", key)
    print(get_dividends_ss(key, 2021))
    print('-------------------------------------------------------------------')
```

```python
def get_stock_price_early(id, year):
    prices = yf.download(id, start="{}-01-01".format(year), end="{}-01-10".format(year), progress=False)
    return prices.iloc[0]['Open'] if len(prices) > 0 else 0
```

```python
get_stock_price_early('STLAP.PA', 2023)
```

```python
data = []
for key in keys:
    data.append([key, get_stock_price_end(key, 2021), get_dividends_ss(key, 2021), get_stock_price_early(key, 2022), get_stock_price_end(key, 2022), get_dividends_ss(key, 2022)])

df = pd.DataFrame(data, columns=['key', 'price_end_2021', 'dividends_2021', 'price_early_2022', 'price_end_2022', 'dividends_2022'])
```

```python
df
```

```python
# fix outlier MT.PA:
df.loc[df['key'] == 'MT.PA', 'dividends_2021'] = .55
```

MT.PA and URW.PA appears to be some kinda outlier.
We put them off the dataframe, though they deserve better treatment.

```python
df = df[df['key'] != 'MT.PA']
df = df[df['key'] != 'URW.PA']
df = df.reset_index(drop=True)
```

```python
df
```

```python
df['yield_2021'] = df['dividends_2021'] / df['price_end_2021']
df['performance_2022'] = (df['dividends_2022'] + df['price_end_2022'] - df['price_early_2022']) / df['price_early_2022']
df
```

## Dogs10

```python
dogs10 = df.sort_values(by='yield_2021', ascending=False, ignore_index=True)[:10]
dogs10
```

```python
def df_performance_2022(df):
    return (df['dividends_2022'].sum() + df['price_end_2022'].sum() - df['price_early_2022'].sum()) / df['price_early_2022'].sum()
```

```python
dogs10_performance_2022 = df_performance_2022(dogs10)
dogs10_performance_2022
```
